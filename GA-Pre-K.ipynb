```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
 population = range(10000)
```

```python
choices = np.random.choice(population,size=100)
sampling = []

for i in range(1,10):
    i = np.mean(np.random.choice(population,size=100))
    sampling.append(i)
sampling
plt.hist(sampling)
plt.title('Ten Samples of size 100')
plt.show()
```

```python
for i in range(1,100):
    i = np.mean(np.random.choice(population,size=100))
    sampling.append(i)
sampling
plt.hist(sampling)
plt.title('Hundred Samples of size 100')
plt.show()
```

_According to `Central Limit Theorem` as the sample size gets larger and larger the distrubition of sampling distribution tends to normal distribution_
